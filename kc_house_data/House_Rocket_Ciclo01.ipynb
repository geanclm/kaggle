# Import

```python
import pandas as pd
from sklearn import linear_model as lm
import numpy as np
```

# Loading dataset

```python
# fonte: https://www.kaggle.com/shivachandel/kc-house-data
```

```python
df = pd.read_csv('kc_house_data.csv')
```

```python
# para resover o erro encontrado no comando "modelo.fit(X,y) informando ausente de dados"
# Obs: dois campos ausentes
# Decisão de exclusão das duas linhas respectivas
# by geanclm in 20/02/2022 at 15h
df = df.dropna(how='any') # apagar linhas com algum dado ausente
```

# Data preparation

```python
X = df.drop(['price','date'],axis=1)
y = df['price'].copy()
```

```python
X
```

```python
y
```

# Model Training

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html
modelo = lm.LinearRegression()
```

```python
modelo.fit(X,y)
```

```python
pred = modelo.predict(X)
```

```python
pred[0:100]
```

# Perfomance Metrics

```python
df1 = df.copy()
```

```python
df1['prediction'] = pred
```

```python
df1['error'] = df1['price']-df1['prediction']
```

```python
df1['error_abs'] = np.abs(df1['error'])
```

```python
df1['error_perc'] = ((df1['price']-df1['prediction'])/df1['price'])*100
```

```python
df1['error_perc_abs'] = np.abs(df1['error_perc'])
```

```python
df1[['id','price','prediction','error','error_abs','error_perc','error_perc_abs']].head()
```

```python
# mean absolute error
mae = np.mean(df1['error_abs'])
print (f'MAE: {mae}')
```

```python
# mean absolute percentage error
mape = np.mean(df1['error_perc_abs'])
print (f'MAPE: {mape}%')
```
